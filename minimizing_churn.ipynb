**Importing Libraries and Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn_data.csv')
```

**Cleaning the Dataset**

```python
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
#checking which columns have NULL values
df.isna().any()
```

```python
#total number of NULL values in each column
df.isna().sum()
```

```python
#removing 4 rows which have NULL values in the age column
df = df[pd.notnull(df['age'])]
```

```python
#as the rewards_earned and credit_score columns have large number of NULL values we will remove them.
df = df.drop(columns=['credit_score','rewards_earned'])
```

```python
df.isna().sum()
```

**Visualizing the Data**

Histogram

```python
df_num = df.drop(columns=['user','churn'])
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Histograms of Numerical Columns',fontsize=20)
for i in range(1,df_num.shape[1]+1):
    plt.subplot(10,3,i)   #6 rows and 5 columns
    f = plt.gca()
    f.set_title(df_num.columns.values[i-1])

    vals = np.size(df_num.iloc[:,i-1].unique())

    plt.hist(df_num.iloc[:,i-1],bins=vals,color='#3F5D7D')

plt.tight_layout(rect = [0,0.03,1,0.95])
```

Pie Charts

```python
#ensuring binary columns are useful to us and can be used in the model (basically are not skewed)

df_bin = df[['housing', 'is_referred', 'app_downloaded',
                    'web_user', 'app_web_user', 'ios_user',
                    'android_user', 'registered_phones', 'payment_type',
                    'waiting_4_loan', 'cancelled_loan',
                    'received_loan', 'rejected_loan', 'zodiac_sign',
                    'left_for_two_month_plus', 'left_for_one_month', 'is_referred']]
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Pie Charts for Binary Columns',fontsize=20)
for i in range(1,df_bin.shape[1]+1):
    plt.subplot(6,3,i)   #6 rows and 5 columns
    f = plt.gca()
    f.set_title(df_bin.columns.values[i-1])

    values = df_bin.iloc[:,i-1].value_counts(normalize=True).values
    index = df_bin.iloc[:,i-1].value_counts(normalize=True).index

    plt.pie(values, labels = index, autopct='%1.1f%%')
    plt.axis('equal')

plt.tight_layout(rect = [0,0.03,1,0.95])
```

Correlation Plot

```python
#good measure of which fields may impact the model a lot
```

```python
df_corr = df.drop(columns=['churn','user','housing','payment_type','zodiac_sign'])   #keeping only numerical columns and removing all others including categorical columns too
```

```python
c = ['red', 'yellow', 'black', 'blue', 'orange','green','violet']
df_corr.corrwith(df['churn']).plot(kind='bar',figsize=(20,10),title='Correlation with Churn Column',fontsize=15,rot=45,grid=True, color=c)
```

Correlation Matrix

```python
sns.set(style='white',font_scale = 2)     #background
corr = df.drop(columns=['user','churn']).corr()    #actual correlation

f,ax = plt.subplots(figsize=(18,15))    #set up matplotlib figure
f.suptitle("Correlation Matrix",fontsize=48)

cmap = sns.diverging_palette(220,10,as_cmap=True)     #set up a custom diverging colormap

sns.heatmap(corr,cmap=cmap,vmax=0.3,center=0,square=True,linewidths=0.5,cbar_kws={"shrink":0.5})
```

```python
#we cannot have highly correlated features 

#ios_user is related to android_user. if it has value 1 ios user has value 0. they are stronly negatively correlated

#similarly web_app_user is a function of two columns web_user and app_user
```
